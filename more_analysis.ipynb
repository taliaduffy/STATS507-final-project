```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

alice = pd.read_csv("alice_in_wonderland/alice_in_wonderland_results.csv")
alice_sents = pd.read_csv("alice_in_wonderland/alice_in_wonderland_sentences.csv")
alice["paragraph"] = alice_sents["paragraph"]
alice.name = "alice"

anne = pd.read_csv("anne_of_green_gables/anne_results.csv")
anne_sents = pd.read_csv("anne_of_green_gables/anne_sentences.csv")
anne["paragraph"] = anne_sents["paragraph"]
anne.name = "anne"

dorian = pd.read_csv("dorian_gray/dorian_gray_results.csv")
dorian_sents = pd.read_csv("dorian_gray/dorian_sentences.csv")
dorian["paragraph"] = dorian_sents["paragraph"]
dorian.name = "dorian"

frankenstein = pd.read_csv("frankenstein/frankenstein_results.csv")
frankenstein_sents = pd.read_csv("frankenstein/frankenstein_sentences.csv")
frankenstein["paragraph"] = frankenstein_sents["paragraph"]
frankenstein.name = "frankenstein"

expectations = pd.read_csv("great_expectations/great_expectations_results.csv")
expectations_sents = pd.read_csv("great_expectations/great_expectations_sentences.csv")
expectations["paragraph"] = expectations_sents["paragraph"]
expectations.name = "expectations"

gatsby = pd.read_csv("great_gatsby/great_gatsby_results.csv")
gatsby_sents = pd.read_csv("great_gatsby/gatsby_sentences.csv")
gatsby["paragraph"] = gatsby_sents["paragraph"]
gatsby.name = "gatsby"

huckleberry = pd.read_csv("huckleberry/huckleberry_results.csv")
huckleberry_sents = pd.read_csv("huckleberry/huckleberry_finn_sentences.csv")
huckleberry["paragraph"] = huckleberry_sents["paragraph"]
huckleberry.name = "huckleberry"

pride = pd.read_csv("pride_and_prejudice/pride_and_prejudice_results.csv")
pride_sents = pd.read_csv("pride_and_prejudice/pride_and_prejudice_sentences.csv")
pride["paragraph"] = pride_sents["paragraph"]
pride.name = "pride"



df = dorian
```

```python
for book in [alice, anne, dorian, frankenstein, expectations, gatsby, huckleberry, pride]:
    perc_neut = (book["label"].value_counts() / len(book))["neutral"]
    print(book.name)
    print(perc_neut)
    print("")
```

```python
emotions = ["neutral", "disgust", "anger", "sadness", "fear", "surprise", "joy"]
colors = ["dimgray", "yellowgreen", "firebrick", "cornflowerblue", "darkorange", "mediumpurple", "gold"]
```

```python
label_freq = df["label"].value_counts()
ordered_emotions = list(label_freq.index)
#color options = ["dimgrey", "gold", "yellowgreen", "cornflowerblue", "mediumpurple", "firebrick", "darkorange"]
ordered_colors = list()
for ordered_emotion in ordered_emotions:
    color_idx = np.where(np.array(emotions) == ordered_emotion)[0][0]
    ordered_colors.append(colors[color_idx])
plt.bar(x = label_freq.index, height = label_freq, color = ordered_colors)
plt.ylabel("Number of sentences")
plt.title("Emotion classification of sentences in the novel")
#plt.savefig("dorian_gray/dorian_gray_barchart.png", bbox_inches = "tight")
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7)  = plt.subplots(7, sharex = True, sharey = True, figsize = (10,5))
axs = (ax1, ax2, ax3, ax4, ax5, ax6, ax7)
emotions = ["neutral", "disgust", "anger", "sadness", "fear", "surprise", "joy"]
cs = ["dimgray", "yellowgreen", "firebrick", "cornflowerblue", "darkorange", "mediumpurple", "gold"]

fig.suptitle("Strength of different emotions throughout novel")
for i in range(7):
    emotion = emotions[i]
    ax = axs[i]
    c = cs[i]
    df_result_par = df[["paragraph", emotion]].groupby("paragraph").agg("mean").reset_index()
    ax.plot((df_result_par[emotion])**2, c)
    ax.set_title(emotion.capitalize(), loc = "left", y = 0.25, x = - 0.1, fontsize = 7)
    ax.set_yticks([0, 0.5, 1])
    ax.tick_params(axis = "y", labelsize = 7)

fig.supxlabel("Paragraph number")
fig.supylabel("Squared Probability")

#plt.savefig("dorian_gray/dorian_gray_linegraph.png", bbox_inches = "tight")
```

```python
# 5 most emotional (least neutral) sentences
most_emotional_sentences = np.argsort(df["neutral"])[0:5]
most_emotional_sentences
```

```python
for text in df["text"].iloc[most_emotional_sentences]: print(text)
```

```python
df_result_par = df.groupby("paragraph").agg("mean", numeric_only = True).reset_index()
```

```python
# 5 most emotional paragraphs 

most_emotional_paragraphs = np.argsort(df_result_par["neutral"])[0:5]
most_emotional_paragraphs
```

```python
df_emotional_paragraphs = df[df["paragraph"].isin(most_emotional_paragraphs)]
for text in df_emotional_paragraphs["text"]: print(text)
```

```python
# Most emotional by max probability for label

max_probs = df.nlargest(n = 5, columns = "prob")
for text in max_probs["text"]: print(text)
```
